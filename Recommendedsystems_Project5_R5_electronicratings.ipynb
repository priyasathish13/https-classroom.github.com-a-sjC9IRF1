```python
ratingscount= q.groupby('productid')['ratings'].count()
ratingscount.head()
TopProduct = df1.groupby(['productid'])['ratings'].mean().reset_index(name='MeanRating')
TopProduct.sort_values('MeanRating', ascending=False).head(20)
```

```python
TopRating_df = df1[df1['ratings'] == 5]
Product_reco = TopRating_df.groupby(['productid']).size().reset_index(name='Counts')
Product_reco.sort_values(by=['Counts'], ascending=False, inplace=True)
Product_reco['Rating_Rank'] = Product_reco['Counts'].rank(ascending= 0)
Product_reco.head(20)
```

```python
from surprise import Dataset,Reader
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df1[['userid', 'productid', 'ratings']], reader)
data
```

```python
from surprise.model_selection import train_test_split
trainset, testset = train_test_split(data, test_size=.3,random_state=100,shuffle=True)
```

```python
from surprise import KNNWithMeans
from surprise import accuracy
from surprise import Prediction
```

```python
algo = KNNWithMeans(k=51, sim_options={'name': 'pearson', 'user_based': False})
algo.fit(trainset)
```
