```python
def Goal_diff_count(laliga):
    print("Team \t : Goal_Diff_Count")
    for row in laliga.itertuples():
        print("%s:%s"%(row.Team,row.GoalsFor-row.GoalsAgainst))
    #print("Team : %s, Goal_diff_count: %s" %(laliga['Team'],laliga["GoalsFor"]-laliga["GoalsAgainst"]))
    print ("Team Max Goal Difference \n", laliga[['Team','Goal_diff_count']][laliga.Goal_diff_count == laliga.Goal_diff_count.max()].to_string(header=False,index=False))
    print ("Team Min Goal Difference \n", laliga[['Team','Goal_diff_count']][laliga.Goal_diff_count == laliga.Goal_diff_count.min()].to_string(header=False,index=False))

    
Goal_diff_count(laliga)
```
